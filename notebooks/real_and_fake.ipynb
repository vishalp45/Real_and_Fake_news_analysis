```python
df["text"] = df["text"].str.lower()
```

```python
df["content"] = df["title"] + " " + df["text"]
```

```python
import re
```

```python
def remove_url(text):
    return re.sub(r"http\S+|www\S+", "", text)
```

```python
df["content"] = df["content"].apply(remove_url)
```

```python
# URLs generally don't contribute to the linguistic content of an article and often introduce noisy, 
# unique tokens into the vocabulary. Removing them helps reduce unnecessary features before TF-IDF vectorization.
```

```python
df["content"].str.contains("<").sum()
```

```python
def remove_html(text):
    return re.sub(r"<.*?>", "", text)
```

```python
df["content"] = df["content"].apply(remove_html)
```

```python
df["content"].str.contains("<").sum()
```

```python
# HTML tags are formatting elements rather than linguistic content. 
# They increase vocabulary size without contributing meaningful information, 
# so I removed them to reduce noise before feature extraction.
```

```python
import string
```

```python
print(string.punctuation)
```

```python


translator = str.maketrans("", "", string.punctuation)
```

```python
df["content"] = df["content"].apply(
    lambda x: x.translate(translator)
)
```

```python
# Punctuation increases the vocabulary without adding much useful information for TF-IDF-based models. 
# Removing it normalizes the text, reduces feature sparsity, and simplifies the representation of words.
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
```

```python
list(stop_words)[:20]
```

```python
def remove_stopwords(text):
    words = text.split()

    filtered_words = [
        word for word in words
        if word not in stop_words
    ]

    return " ".join(filtered_words)
```

```python
df["content"] = df["content"].apply(remove_stopwords)
```

```python
# Stopwords occur very frequently but usually contribute little to distinguishing between classes. 
# Removing them reduces the vocabulary size and focuses the model on more informative words.
# However, I retained important negation words such as 'not', 'no', and 'nor' because they significantly affect sentence meaning
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
print(stemmer.stem("playing"))
```

```python
def stem_text(text):
    words = text.split()

    stemmed_words = [
        stemmer.stem(word)
        for word in words
    ]

    return " ".join(stemmed_words)
```

```python
df["content"] = df["content"].apply(stem_text)
```

```python
# Since this project uses TF-IDF with traditional machine learning algorithms,
# I chose Porter Stemming because it's computationally efficient and effectively reduces vocabulary size.
# Although stemming may produce non-dictionary words,
# that's acceptable because TF-IDF treats words as tokens rather than relying on their grammatical correctness.
```

```python
words = [
    "playing",
    "played",
    "plays",
    "studies",
    "studying",
    "happily",
    "football",
    "connections",
    "universities"
]

for word in words:
    print(word, " --> ", stemmer.stem(word))
```

```python
# Bag of Words only counts word occurrences and treats all words equally. 
# TF-IDF improves upon it by reducing the importance of very common words while giving higher 
# weight to words that are rare across documents but important within a specific document. 
#     This often leads to better text classification performance.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_features=5000,
    min_df=2,
    ngram_range=(1,2)
)

X = vectorizer.fit_transform(df["content"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    df["label"],
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)
```

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.value_counts())

print(y_test.value_counts())
```

```python
# Stratified splitting preserves the class distribution in both the training and testing sets.
# This ensures that both sets remain representative of the original dataset and prevents biased evaluation caused by uneven class proportions."
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

print("Accuracy :", accuracy_score(y_test, y_pred))

print("Precision :", precision_score(y_test, y_pred))

print("Recall :", recall_score(y_test, y_pred))

print("F1 Score :", f1_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Fake","Real"],
    yticklabels=["Fake","Real"]
)

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.title("Confusion Matrix")

plt.show()
```

```python
def plot_confusion_matrix(model, X_test, y_test, model_name):
    
    from sklearn.metrics import confusion_matrix
    import seaborn as sns
    import matplotlib.pyplot as plt

    predictions = model.predict(X_test)

    cm = confusion_matrix(y_test, predictions)

    plt.figure(figsize=(7,6))

    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=["Fake","Real"],
        yticklabels=["Fake","Real"]
    )

    plt.xlabel("Predicted Label")
    plt.ylabel("Actual Label")
    plt.title(f"Confusion Matrix - {model_name}")

    plt.show()
```
